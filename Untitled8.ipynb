```python
import matplotlib.pyplot as plt
import matplotlib as mp
import numpy as np

year = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
rainfall = [68.3, 77.98, 98.93, 92.2, 100.68, 96.8, 98.27, 92.42, 79.48, 101.3, 93.02, 87.89, 78.1, 65.13, 80.34, 98.76, 78.2]

#plot the bar graph
plot = plt.bar(year, rainfall )

#Add the data value on the head of the bar
for value in plot:
    height = value.get_height ()
   
#Add labels and title
plt.title("Rainfall in India in different years")
plt.xlabel("year")
plt.ylabel("rainfall in MM")

#Display the graph on the screen
plt.show()
    

```
